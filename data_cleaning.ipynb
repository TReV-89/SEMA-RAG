```python
import numpy as np
import pandas as pd

sema_df_1 = pd.read_csv('SEMA_DATA/kabalagala-english-transcriptions - kabalagala-english-transcriptions.csv')
sema_df_2 = pd.read_csv('SEMA_DATA/kabalagala-luganda-transcriptions - kabalagala-luganda-transcriptions.csv')
sema_df_3 = pd.read_csv('SEMA_DATA/KampalaHospitalData - KampalaHospitalData.csv')
sema_df_4 = pd.read_csv('SEMA_DATA/kasangatienglish-transcriptions - kasangati-english.csv')
sema_df_5 = pd.read_csv('SEMA_DATA/kasangatiluganda-transcriptions - kasangatiluganda-transcriptions.csv')
sema_df_6 = pd.read_csv('SEMA_DATA/KiraRoadPoliceData - overalloutput.csv')
sema_df_7 = pd.read_csv('SEMA_DATA/KisuguHospitalData - KisuguHospitalData.csv')
sema_df_8 = pd.read_csv('SEMA_DATA/KiswaHospitalData - KiswaHospitalData.csv')
sema_df_9 = pd.read_csv('SEMA_DATA/komamboga-english-transcriptions - komamboga-english-transcriptions.csv')
sema_df_10 = pd.read_csv('SEMA_DATA/komamboga-luganda-transcriptions - komamboga-luganda-transcriptions.csv')
sema_df_11 = pd.read_csv('SEMA_DATA/oldklaenglish-transcriptions - oldklaenglish-transcriptions.csv')
sema_df_12 = pd.read_csv('SEMA_DATA/oldklaluganda-transcriptions - oldklaluganda-transcriptions.csv')

print("Current columns in sema_df_12:")
print(sema_df_12.columns.tolist())


sema_df_4 = sema_df_4.drop(columns=['sentiment-analysis chatgpt'])
sema_df_5 = sema_df_5.drop(columns=['sentiment analysis-chatgpt','corrected translation'])
sema_df_11 = sema_df_11.drop(columns=['sentiment-analysis chatgpt'])
sema_df_12 = sema_df_12.drop(columns=['sentiment analysis - chatgpt', 'correction.1'])
```

```python
sema_df_1 = sema_df_1.dropna()
sema_df_2 = sema_df_2.dropna()
sema_df_3 = sema_df_3.dropna()  
sema_df_4 = sema_df_4.dropna()
sema_df_5 = sema_df_5.dropna()
sema_df_6 = sema_df_6.dropna()
sema_df_7 = sema_df_7.dropna()
sema_df_8 = sema_df_8.dropna()
sema_df_9 = sema_df_9.dropna()
sema_df_10 = sema_df_10.dropna()
sema_df_11 = sema_df_11.dropna()
sema_df_12 = sema_df_12.dropna()
```

```python
import os

# Create exports directory if it doesn't exist
export_dir = 'SEMA_DATA_exports'
os.makedirs(export_dir, exist_ok=True)

# Dictionary mapping dataframes to their export names
dataframes = {
    'kabalagala_english': sema_df_1,
    'kabalagala_luganda': sema_df_2,
    'kampala_hospital': sema_df_3,
    'kasangati_english': sema_df_4,
    'kasangati_luganda': sema_df_5,
    'kira_road_police': sema_df_6,
    'kisugu_hospital': sema_df_7,
    'kiswa_hospital': sema_df_8,
    'komamboga_english': sema_df_9,
    'komamboga_luganda': sema_df_10,
    'oldkla_english': sema_df_11,
    'oldkla_luganda': sema_df_12
}

# Export to different formats
for name, df in dataframes.items():
    # CSV export
    excel_path = os.path.join(export_dir, f'{name}.xlsx')
    df.to_excel(excel_path, index=False)
    

print(f"All files exported to {export_dir}")
```
